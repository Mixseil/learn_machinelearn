```python
import numpy as np    # для функции линейной алгебры
import pandas as pd   # для таких объектов как датафрейм и серия, храниние структурированной табличной информации
import matplotlib.pyplot as plt # для визуализации
```

```python
x = pd.read_csv('https://github.com/koroteevmv/ML_course/raw/refs/heads/main/ML1.1%20linear%20regression/data/x.csv', index_col=0)['0']
y = pd.read_csv('https://github.com/koroteevmv/ML_course/raw/refs/heads/main/ML1.1%20linear%20regression/data/y.csv', index_col=0)['0']
```

```python
class Model(object):
  """Модель парной линейной регрессии у = b0 + b1 * x"""
  def __init__(self, max_iter=50): # объявляем 2 параметра модели
      self.b0 = 1
      self.b1 = 1
      self.max_iter = max_iter

  def predict(self, x): # Прямое действие модели, предсказание значений, передаём значения атрибута модели и ищем y который может быть у модели
      return self.b0 + self.b1 * x

  def error(self, x, y):
    # """J = y_pred - y_true"""
      return sum((self.predict(x) - y)**2) / (2 * len(y))

# из урока: фишка библеотеки numpy - векторизация значений x может быть как числом так и серией npArray, y будет таким же, либо один либо вектор по этому выше импользуем сумму и делим на количество

  def fit(self, x, y): # сам градиентный спуск
      alpha = 0.1
      errors = []
      errors = [self.error(x, y), ]
      for _ in range(self.max_iter):
        dJ0 = sum(self.predict(x) - y) / len(y)
        dJ1 = sum((self.predict(x) - y) * x) / len(y)
        self.b0 -= alpha * dJ0
        self.b1 -= alpha * dJ1

        new_err = self.error(x, y)
        old_err = errors[-1]
        if new_err > old_err:
          alpha /= 2
          self.__init__()

        if old_err - new_err < 1e-5:
          break
        errors.append(new_err)
      else:
        print("warning")

      return errors
```

```python
model = Model(20)

errors = model.fit(x, y)
plt.plot(errors, 'g')
```

```python
model.fit(x, y)
print("Начальная ошибка", model.error(x, y)) # зависит от начальных b0 и b1 надо найти значения чтоб ошибка была минимальной
print("Конечная ошибка:", model.error(x, y))
```

```python
plt.figure()
plt.scatter(x, y)

x_pred = np.linspace(-2.5, 2, 100)
y_pred = model.predict(x_pred)
plt.plot(x_pred, y_pred, 'r')
plt.show()
```
